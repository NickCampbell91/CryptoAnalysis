```python
import pandas as pd 

raw_data = pd.read_csv("consolidated_coin_data.csv")
raw_data.head()
```

```python
df = raw_data.dropna()
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df["Open"] = pd.to_numeric(df["Open"], errors = "coerce")
df["High"] = pd.to_numeric(df["High"], errors = "coerce")
df["Low"] = pd.to_numeric(df["Low"], errors = "coerce")
df["Close"] = pd.to_numeric(df["Close"], errors = "coerce")
df["Volume"] = pd.to_numeric(df["Volume"], errors = "coerce")
df["Market Cap"] = pd.to_numeric(df["Market Cap"], errors = "coerce")

df.head()
```

```python
df.dtypes
```

```python
df.plot(x="Date",y="Open")
```

```python
df_tezos = df[df["Currency"] == "tezos"]
df_tezos.plot(x="Date",y="Open")
```

```python
df_binance = df[df["Currency"] == "binance-coin"]
df_binance.plot(x="Date",y="Open")
```

```python
df_eos = df[df["Currency"] == "eos"]
df_eos.plot(x="Date",y="Open")
```

```python
df_bitcoin = df[df["Currency"] == "bitcoin"]
df_bitcoin.plot(x="Date",y="Open")
```

```python
df_tether = df[df["Currency"] == "tether"]
df_tether.plot(x="Date",y="Open")
```
